```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, value):
        newnode = Node(value)
        if self.root is None:
            self.root = newnode
        temp = self.root
        if newnode.val < temp.val:
            if temp.value == newnode.value:
                return False
            if newnode.value < temp.value:
                if temp.left is None:
                    temp.left = newnode
                    return True
                temp = temp.left
            else:
                if temp.right is None:
                    temp.right = newnode
                    return True
                temp = temp.right
                
```
