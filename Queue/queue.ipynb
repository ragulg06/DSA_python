```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Queue:
    def __init__(self):
        self.first = None
        self.last = None
        self.length = 0

    def enqueue(self, value):
        newnode = Node(value)
        if self.first is None:
            self.first = newnode
            self.last = newnode
        else:
            self.last.next = newnode
            self.last = newnode
        self.length += 1
        self.print_queue()
        return True

    def dequeue(self):
        if self.length == 0:
            return None
        temp = self.first
        if self.length == 1:
            self.first, self.last = None, None
        else:
            self.first = self.first.next
            temp.next = None
        self.length -= 1
        self.print_queue()
        return temp
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(f"{temp.value}-> ", end="")
            temp = temp.next
        print(f"length = {self.length}")
```

```python
q = Queue()
for i in range(1, 6):
    q.enqueue(i)
```
